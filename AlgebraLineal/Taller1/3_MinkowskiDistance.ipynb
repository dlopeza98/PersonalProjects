## 3.A) Plotting Minkowski Distance

### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
```

### Trying function for diferent values of `p`

```python
p_values = [1,2,3,4,5,6,7,1000]  # Infinity represented as np.inf

# Create a figure to hold the plots
plt.figure(figsize=(6, 6))

# Iterate through different values of p
for p in p_values:
    # Generate theta values
    theta = np.linspace(0, 2 * np.pi, 1000)
    radius = 1


    factor_r = (radius / ((np.abs(np.cos(theta))**p + np.abs(np.sin(theta))**p) **(1/p)))

    # Calculate x and y coordinates from the polar coordinates
    x = factor_r * np.cos(theta)
    y = factor_r * np.sin(theta)

    # Plot the perimeter with the current p value
    plt.plot(x, y, label=f'p={p}')

# Set labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Plot of Minkowski Distance for different values of p')
plt.legend()
plt.grid(True)
plt.axis('equal')  # Equal aspect ratio
plt.show()
```

### Creating function to plot Minkowski Distance

```python
def plot_minkowski_distance(p,r):
    theta = np.linspace(0, 2 * np.pi, 1000)
    radius=r
    factor_r = (radius / ((np.abs(np.cos(theta))**p + np.abs(np.sin(theta))**p) **(1/p)))
    x = factor_r * np.cos(theta)
    y = factor_r * np.sin(theta)

    # Plot the perimeter with the current p value
    plt.figure(figsize=(4,4 ))
    plt.plot(x, y, label=f'p={p}, r={r}')
    
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Plot of Minkowski Distance with p={p}, r={r}')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')  # Equal aspect ratio
    plt.show()
```

```python
# Create sliders for p and r
#p_slider = widgets.FloatSlider(value=1.0, min=0.1, max=10.0, step=0.1, description='p:')
p_slider = widgets.BoundedFloatText(value=1.0, description='p:',min=0.1,step=0.1,max=1000)
r_slider = widgets.FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='r:')

# Create an output widget to display the plot
plot_output = widgets.Output()

# Function to update the plot when sliders change
def update_plot(change):
    p = p_slider.value
    r = r_slider.value
    with plot_output:
        clear_output(wait=True)  # Clear the previous output without clearing the whole cell
        plot_minkowski_distance(p, r)

# Attach the update function to the slider values
p_slider.observe(update_plot, names='value')
r_slider.observe(update_plot, names='value')

# Display the widgets
some="Modify the parameters to see the plot"
display(some,p_slider, r_slider, plot_output)
```

## 3.B) Create random Vector with Minkowski distance less than 0.1 from origin

#### Importing Library

```python
import random
```

### Creating function to generate random vectors

The random vector will be of lenght `n` and the Minkowski Distance `p` with the origin is less than `alpha.`

```python
def random_vector(n, alpha, p):
    """
    n= lenght of vector
    p= value for Minkowski Distance
    alpha= Minkowski Distance with origin
    """
    # Se genera un vector de tamao n con nmeros aleatorios en el rango de 0 a 200
    random_vector = np.array([random.uniform(0, 100) for _ in range(n)])
    norm = np.linalg.norm(random_vector, ord=p)
    # Se generan los vectores de tamao n, cuya distancia de Minkowoski de norma p al origen sea menor que alpha
    a = (random_vector/norm)*alpha
    return print(a)
```

```python
n_slider = widgets.BoundedIntText(value=10, description='n:',min=1,step=1,max=10000)
alpha_slider = widgets.BoundedFloatText(value=0.1, description='alpha:',min=0,step=0.01,max=0.1)
p_slider = widgets.BoundedFloatText(value=1.0, min=0.1, max=1000, step=0.1, description='p:')

# Create an output widget to display the plot
plot_output_2 = widgets.Output()

# Function to update the plot when sliders change
def update_plot(change):
    n = n_slider.value
    p = p_slider.value
    alpha = alpha_slider.value
    with plot_output_2:
        clear_output(wait=True)  # Clear the previous output without clearing the whole cell
        random_vector(n, alpha, p)

# Attach the update function to the slider values
p_slider.observe(update_plot, names='value')
n_slider.observe(update_plot, names='value')
alpha_slider.observe(update_plot, names='value')

# Display the widgets
some="Modify the parameters to see the vector"
display(some,p_slider, n_slider, alpha_slider,plot_output_2)
```
