# Variables Aleatorias

## Importando liberías

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm
```

## Distribución Normal

### usando Numpy

```python
# distribución normal aleatoria
N=10000
x_normal=np.random.normal(scale=2,loc=1, size=N) #scale = std, loc = media, size= número de datos
```

```python
plt.title("Datos simulados con una normal")
plt.hist(x_normal,bins=100)
```

### usando Scipy

Para calcular probabilidad especifica de un número

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html

pdf está asociada a función distribución

ppf está asociado al valor inverso

cdf está asociado al la función de distribución acumulado

```python
norm.cdf(0,loc=1,scale=0.7) #Probabilidad de que X sea menor o igual a 0
```

```python
# la más usada en términos de inferencia estadística

print("Cola izquierda ",norm.ppf(0.05))  # Es el valor que indica que la probabilidad acumulada hasta ese punto es 0.05 para la cola izquiera de la distribución

print("Cola derecha ",norm.ppf(0.95)) # Dado que es una distribución simétrica, si queremos la cola derecha, solo le quitamos el valor negativo
```

## Función exponencial
https://numpy.org/doc/stable/reference/random/generated/numpy.random.exponential.html

```python
x_exponencial=np.random.exponential(scale=1/10,size=10000)
```

```python
plt.title("Datos simulados con una exponencial")
plt.hist(x_exponencial,bins=100)
```
